```python
df1[' "Test2"'] = pd.to_numeric(df1[' "Test2"'], errors='coerce')
df1[' "Test3"'] = pd.to_numeric(df1[' "Test3"'], errors='coerce')
df1[' "Test4"'] = pd.to_numeric(df1[' "Test4"'], errors='coerce')
df1[' "Final"'] = pd.to_numeric(df1[' "Final"'], errors='coerce')
```

```python
average_numeric=df1[' "Test2"']*0.2+ df1[' "Test3"']*0.2+ df1[' "Test4"']*0.2+ df1[' "Final"']*0.4
average_numeric
```

```python
df1["'Numeric_Grade'"]=average_numeric
df1
```

```python
df1.columns = df1.columns.str.strip().str.replace('"', '').str.replace("'", '')
df1.columns
```

#Query operations on a DataFrame


```python
df2 = pd.DataFrame({'Year': [1776, 1867, 1821], 'Pop': [328, 38, 126], 'GDP': [20.5, 1.7, 1.22]}, index=['USA', 'CANADA', 'MEXICO'])
df2
```

```python
#Which countries have a population of more than 50 million?
df2['Pop']
```

```python
df2['Pop']>50
```

```python
df2[df2['Pop']>50]
```

```python
#Which countries wereborn before 1800
df2['Year']
```

```python
df2['Year']>1800
```

```python
df2[df2['Year']>1800]
```
